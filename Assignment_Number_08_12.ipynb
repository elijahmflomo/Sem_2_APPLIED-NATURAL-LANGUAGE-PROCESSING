**Goal:**
Rank important words in a document using the TextRank, which is based on PageRank.

**Step 2: Import Libraries**

```python
import nltk
import networkx as nx
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
import string
```

```python
# Download required datasets

nltk.download('punkt')
nltk.download('stopwords')
```

**Step 3: Input Text (Sample News Text)**

```python
text = """
Natural language processing helps computers understand human language.
Modern NLP techniques improve search engines and recommendation systems.
"""
```

**Step 4: Text Preprocessing**

Steps:

Convert to lowercase

Tokenize words

Remove stopwords and punctuation

```python
stop_words = set(stopwords.words('english'))

tokens = word_tokenize(text.lower())

words = [word for word in tokens if word not in stop_words and word not in string.punctuation]

print("Filtered Words:", words)
```

**Step 5: Construct Word Co-occurrence Graph**

We connect words appearing within a window size = 2.

```python
window_size = 2

graph = nx.Graph()

for i in range(len(words)):
    for j in range(i+1, i+window_size):
        if j < len(words):
            graph.add_edge(words[i], words[j])


print(graph)
```

Graph meaning:

word → node

co-occurring words → edge

**Step 6: Apply TextRank (PageRank)**

```python
scores = nx.pagerank(graph)

print("Word Scores:", scores)
```

**Step 7: Display Word Rankings**

```python
print("Word Importance Scores:\n")

for word, score in sorted(scores.items(), key=lambda x: x[1], reverse=True):
    print(word, ":", round(score,4))
```

**Step 8: Extract Top Keywords**

```python
top_keywords = sorted(scores, key=scores.get, reverse=True)[:5]

print("\nTop Keywords:", top_keywords)
```

**Conclusion**

*The TextRank algorithm was implemented using a word co-occurrence graph. Each word was treated as a node and edges were created between words appearing within a fixed window size. The PageRank algorithm was applied to compute importance scores for each word. The highest ranked words were selected as keywords, representing the most important terms in the document.*

**Q2: Extract Top 10 Keywords using TextRank**

**Goal**

*From the ranking scores produced by the TextRank, extract the top 10 most important words in the document.*

*TextRank works similarly to PageRank, ranking nodes (words) in a graph. *

**Step 1: Complete Python Script**

This script includes:

*preprocessing*

*graph construction*

*TextRank scoring*

*keyword extraction*

```python
import nltk
import networkx as nx
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
import string

# Sample document
text = """
Natural language processing helps computers understand human language.
Modern NLP techniques improve search engines and recommendation systems.
"""

# Download resources (run once)
nltk.download('punkt')
nltk.download('stopwords')

# Step 1: Preprocess text
stop_words = set(stopwords.words('english'))

tokens = word_tokenize(text.lower())

words = [word for word in tokens
         if word not in stop_words and word not in string.punctuation]

# Step 2: Build word co-occurrence graph
window_size = 2
graph = nx.Graph()

for i in range(len(words)):
    for j in range(i+1, i+window_size):
        if j < len(words):
            graph.add_edge(words[i], words[j])

# Step 3: Apply TextRank (PageRank)
scores = nx.pagerank(graph)

# Step 4: Extract Top 10 keywords
top_keywords = sorted(scores.items(), key=lambda x: x[1], reverse=True)[:10]

print("Top 10 Keywords with Scores:\n")

for word, score in top_keywords:
    print(word, ":", round(score,4))
```

**Step 3: Interpretation**

*The top-ranked words represent the most important concepts in the document. Words such as language, recommendation, engines, search, and NLP receive higher scores because they appear frequently and are strongly connected with other important words in the text.*

*The TextRank algorithm ranks these words higher based on their relationships in the word co-occurrence graph, indicating that they are central to the document's topic. These keywords summarize the main theme of the article and can be used for search indexing, document summarization, and recommendation systems.*

**Q3: Compare Keywords from Two Documents**

**Goal**

*Apply the TextRank program to two different documents.*

*Extract the top keywords from each document.*

*Compare the results and discuss how NLTK preprocessing improves keyword relevance.*

**Step 1: Example Documents**

```python
doc1 = """
Natural language processing enables computers to understand human language.
Modern NLP techniques improve search engines and recommendation systems.
"""

doc2 = """
Machine learning algorithms analyze large datasets to discover patterns.
These models are widely used in data science, artificial intelligence, and predictive analytics.
"""
```

**Step 2: Python Program**

This script:

*preprocesses text using NLTK*

*builds a word graph*

*applies TextRank*

*extracts top keywords*

```python
import nltk
import networkx as nx
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
import string

nltk.download('punkt')
nltk.download('stopwords')

stop_words = set(stopwords.words('english'))

def textrank_keywords(text):

    tokens = word_tokenize(text.lower())

    words = [w for w in tokens if w not in stop_words and w not in string.punctuation]

    graph = nx.Graph()
    window_size = 2

    for i in range(len(words)):
        for j in range(i+1, i+window_size):
            if j < len(words):
                graph.add_edge(words[i], words[j])

    scores = nx.pagerank(graph)

    top_words = sorted(scores.items(), key=lambda x: x[1], reverse=True)[:10]

    return top_words


keywords_doc1 = textrank_keywords(doc1)
keywords_doc2 = textrank_keywords(doc2)

print("Document 1 Keywords:")
for w,s in keywords_doc1:
    print(w, ":", round(s,4))

print("\nDocument 2 Keywords:")
for w,s in keywords_doc2:
    print(w, ":", round(s,4))
```

**Step 4: Comparison**

Document	Main Topic	Extracted Keywords

*Doc1	Natural Language Processing	language, nlp, search, engines*

*Doc2	Machine Learning	machine, learning, datasets, analytics*

TextRank successfully identifies topic-specific keywords for each document.
